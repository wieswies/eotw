## 1. List corrected classfications
After the embeddings of cut-out news images were labelled, they were manually inspected*. There are three types of misclassifications:
- a. Misclassifications (FP): Unknowns that got the label of a politician
- b. Missed classifications (FN): Politicians that got the Unknown label
- c. Mistaken classifications (FP/FN): Politicians mistaken for each other </br></br>

*The manual inspection consisted of:
1. Copy the whole folder structure with the classifications into a new folder
2. Manually go through the folders to check whether the model classifications were correct
3. Drag and drop the misclassified images, each identified and traceable by their filename, to the folder they belong to
4. Copy the misclassifications in the folder belonging to their misclassification type, i.e. one of the following: '_Misclassifications_FP', '_Missed_Classifications_FN', '_Mistaken_Classifications_FPFN'
5. Go over the corrected folders another two times to minimize human errors
6. Take track of the final, corrected label, by adding it as a column to the data by means of looping through the corrected folder structure

```python
nos_nu_uncorrected = pd.read_parquet('datasets/embeddings/nos_nu_embeddings_labelled_uncorrected_t1t2.parquet')
```

```python
print(f'The columns in the NOS/NU output (uncorrected) after model classification process contains the following columns:\n{nos_nu_uncorrected.columns.tolist()}')
print(f'\nThis dataset (including both NOS and NU images)is of the following shape: {nos_nu_uncorrected.shape}')
```

Set directories for the classifications (two thresholds values were explored for SVM classification, t1 (0.4) and t2 (0.7))

```python
nos_clf_folder_t1 = 'datasets/images/svc_classification_corrected/NOS_t1'
nos_clf_folder_t2 = 'datasets/images/svc_classification_corrected/NOS_t2'

nu_clf_folder_t1 = 'datasets/images/svc_classification_corrected/NU_t1'
nu_clf_folder_t2 = 'datasets/images/svc_classification_corrected/NU_t2'
```

### a. Correct labels after manual folder inspection
#### i. Check for human classification errors by inspecting the corrections for both thresholds

```python
from methods_classification import list_corrected_classes

df_nos_t1 = list_corrected_classes('NOS', 't1', nos_clf_folder_t1)
df_nos_t2 = list_corrected_classes('NOS', 't2', nos_clf_folder_t2)
df_nu_t1 = list_corrected_classes('NU', 't1', nu_clf_folder_t1)
df_nu_t2 = list_corrected_classes('NU', 't2', nu_clf_folder_t2)
```

```python
# Create dataframes with both true labels to inspect human error
df_nos = df_nos_t1.merge(
    df_nos_t2[['dataset', 'id', 'image_filename', 'true_label_t2']],
    on=['dataset', 'id', 'image_filename'],
    how='left'
)
df_nu = df_nu_t1.merge(
    df_nu_t2[['dataset', 'id', 'image_filename', 'true_label_t2']],
    on=['dataset', 'id', 'image_filename'],
    how='left'
)
```

```python
if not df_nos['true_label_t1'].equals(df_nos['true_label_t2']):
    print('Classification error, check t1/t2 true values for NOS.\n')
    print(f'The rows for which true values do not match are:\n{df_nos.query('true_label_t1 != true_label_t2')}\n\n')
else: print('Nu human classification errors for the NOS dataset, true values were correctly found and are identical for the two classification thresholds.\n')

if not df_nu['true_label_t1'].equals(df_nu['true_label_t2']):
    print('Classification error, check t1/t2 true values for NU.\n')
    print(f'The rows for which true values do not match are:\n{df_nu.query('true_label_t1 != true_label_t2')}\n\n')
else: print('Nu human classification errors for the NU dataset, true values were correctly found and are identical for the two classification thresholds.')
```

In case there were entries for which true_value_t1 and true_value_t2 did not match, this means that some misclassifications were missed in the correction of either t1 or t2 correction rounds. The uncomplying entries need another round of scrutiny to find out which classification is correct, and to be adjusted accordingly in the folder structure. </br>
If that is the case, re-run the script to check if all the classifications are right.

#### ii. Finalize dataset with true values for both NOS and NU

```python
df_nos['true_label'] = df_nos['true_label_t1']
df_nos = df_nos.drop(columns = ['true_label_t1', 'true_label_t2'])

df_nu['true_label'] = df_nu['true_label_t1']
df_nu = df_nu.drop(columns = ['true_label_t1', 'true_label_t2'])
```

### b. Join NOS/NU corrected labels

```python
nos_nu_labelled_embeddings_corrected = pd.concat([df_nos, df_nu], ignore_index=True)
```

```python
print(f'The columns in the NOS/NU output (corrected) after model classification process and manual correction contains the following columns:\n{nos_nu_labelled_embeddings_corrected.columns.tolist()}')
print(f'\nThis dataset (including both NOS and NU images)is of the following shape: {nos_nu_labelled_embeddings_corrected.shape}')
```

### c. Merge data to get classification metrics and insight into visual coverage of politicians for theh different news media

```python
nos_nu_corrected = nos_nu_uncorrected.merge(
    nos_nu_labelled_embeddings_corrected[['dataset', 'id', 'image_filename', 'true_label']],
    on=['dataset', 'id', 'image_filename'],
    how='left'
)
```

```python
print(f'The columns in the merged NOS/NU corrected data are:\n{nos_nu_corrected.columns.tolist()}')
```

```python
predicted_labels_t1 = nos_nu_corrected['pred_label_t1']
predicted_labels_t2 = nos_nu_corrected['pred_label_t2']

true_labels = nos_nu_corrected['true_label']
```

```python
cm_t1 = confusion_matrix(true_labels, predicted_labels_t1, labels=classes)
cm_t2 = confusion_matrix(true_labels, predicted_labels_t2, labels=classes)
```

```python
predicted_labels_t1t2 = [{'threshold': 0.4, 'labels': predicted_labels_t1, 'cm': cm_t1}, {'threshold': 0.7, 'labels': predicted_labels_t2, 'cm': cm_t2}]
```

### a. Confusion matrices for both thresholds

```python
# Compute the confusion matrices 
import matplotlib.colors as mcolors

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
norm = mcolors.LogNorm(vmin=1, vmax=max(cm_t1.max(), cm_t2.max()))

sns.heatmap(cm_t1, annot=True, fmt='d', cmap='Blues', ax=axes[0], norm=norm, cbar_kws={'label': 'Count'}, xticklabels=classes, yticklabels=classes)
axes[0].set_xlabel('Predicted', fontdict=axis_font)
axes[0].set_ylabel('True', fontdict=axis_font)
axes[0].set_title(f'a. CM for threshold t1 at p < 0.4', fontdict=title_font)

sns.heatmap(cm_t2, annot=True, fmt='d', cmap='Blues', ax=axes[1], norm=norm, cbar_kws={'label': 'Count'},  xticklabels=classes, yticklabels=classes)
axes[1].set_xlabel('Predicted', fontdict=axis_font)
axes[1].set_ylabel('True', fontdict=axis_font)
axes[1].set_title(f'b. CM for threshold t2 at p < 0.7', fontdict=title_font)
plt.tight_layout()
plt.savefig('../z__visual-outputs-matplotlib/confusion_matrices_SVM_t1t2.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
# Confusion Matrices
cm_t1 = confusion_matrix(true_labels, predicted_labels_t1, labels=classes)
cm_t2 = confusion_matrix(true_labels, predicted_labels_t2, labels=classes)

# Normalizes
cm_t1_normalized = cm_t1.astype('float') / cm_t1.sum(axis=1)[:, np.newaxis]
cm_t2_normalized = cm_t2.astype('float') / cm_t2.sum(axis=1)[:, np.newaxis]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Normalize colors for logarithmic scaling in original matrices
norm = mcolors.LogNorm(vmin=1, vmax=max(cm_t1.max(), cm_t2.max()))

# Original CM
sns.heatmap(cm_t1, annot=True, fmt='d', cmap='Blues', ax=axes[0, 0], norm=norm, cbar_kws={'label': 'Count'}, xticklabels=classes, yticklabels=classes)
axes[0, 0].set_xlabel('Predicted', fontdict=axis_font)
axes[0, 0].set_ylabel('True', fontdict=axis_font)
axes[0, 0].set_title('a. Confusion Matrix for t = 0.4\n', fontdict=title_font)

sns.heatmap(cm_t2, annot=True, fmt='d', cmap='Blues', ax=axes[0, 1], norm=norm, cbar_kws={'label': 'Count'}, xticklabels=classes, yticklabels=classes)
axes[0, 1].set_xlabel('Predicted', fontdict=axis_font)
axes[0, 1].set_ylabel('True', fontdict=axis_font)
axes[0, 1].set_title('b. Confusion Matrix for t = 0.7\n', fontdict=title_font)

# Format normalized values to show blank for 0 and rounded values without leading zero
def fmt_func(value):
    if value == 0:
        return ""
    if value == 1.00:
        return 1
    if value <=0.004:
        return "<.01"
    else:
        return f"{value:.2f}".lstrip('0')

# Plot the normalized confusion matrices
sns.heatmap(cm_t1_normalized, annot=True, fmt='', cmap='Purples', ax=axes[1, 0], 
            cbar_kws={'label': 'Proportion'}, xticklabels=classes, yticklabels=classes,
            annot_kws={"size": 8})

# Apply the custom format to annotations
for text in axes[1, 0].texts:
    value = float(text.get_text())
    text.set_text(fmt_func(value))

axes[1, 0].set_xlabel('Predicted', fontdict=axis_font)
axes[1, 0].set_ylabel('True', fontdict=axis_font)
axes[1, 0].set_title('c. Normalized Confusion Matrix for t = 0.4\n', fontdict=title_font)

sns.heatmap(cm_t2_normalized, annot=True, fmt='', cmap='Purples', ax=axes[1, 1], 
            cbar_kws={'label': 'Proportion'}, xticklabels=classes, yticklabels=classes,
            annot_kws={"size": 8})

# Apply the custom format to annotations
for text in axes[1, 1].texts:
    value = float(text.get_text())
    text.set_text(fmt_func(value))

axes[1, 1].set_xlabel('Predicted', fontdict=axis_font)
axes[1, 1].set_ylabel('True', fontdict=axis_font)
axes[1, 1].set_title('d. Normalized Confusion Matrix for t = 0.7\n', fontdict=title_font)

plt.tight_layout()
plt.savefig('../z__visual-outputs-matplotlib/confusion_matrices_SVM_t1t2_normalized.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

### Group per article id and remove duplicate labels

```python
nos_nu_corrected_only_politicians = nos_nu_corrected[nos_nu_corrected['true_label'] != 'Unknown']
print(f'Out of the {nos_nu_corrected.shape[0]} classified cut-out faces from news images, {nos_nu_corrected_only_politicians.shape[0]} depicted one of the politicians of interest.\nHowever, some duplicate cut-outs occurred in the data, which must be removed to prevent a biased analysis.')
print("This is done by checking whether for a single news image, identified by article id, a single politician's name\noccurs more than once in the 'true_value' classification label.")
```

```python
nos_nu_corrected_only_politicians_no_duplicates = nos_nu_corrected_only_politicians.copy()
nos_nu_corrected_only_politicians_no_duplicates = nos_nu_corrected_only_politicians_no_duplicates.sort_values(by=['id', 'true_label'])
nos_nu_corrected_only_politicians_no_duplicates = nos_nu_corrected_only_politicians_no_duplicates.drop_duplicates(subset=['id', 'true_label'], keep='first')
nos_nu_corrected_only_politicians_no_duplicates = nos_nu_corrected_only_politicians_no_duplicates.reset_index(drop=True)
```

```python
print(f'Out of the {nos_nu_corrected_only_politicians.shape[0]} classified images, {nos_nu_corrected_only_politicians.shape[0]-nos_nu_corrected_only_politicians_no_duplicates.shape[0]} duplicates were found.\nAfter removing these duplicates, the total number of political images is {nos_nu_corrected_only_politicians_no_duplicates.shape[0]}.')
```

### Group by article id in preparation of merger with pre-processed article data
In order to merge the classifications with the article (meta)data in the next step of the project, to facilitate multi-modal analysis, </br>
the only required data is to have the (list of) politician(s) per article id, hence article main image. </br>
Therefore, it is needed to group the classifications by article id and join the 'true_label' values into one list, </br>
such that it is clear whether politicians occurred in the same image together.

```python
nos_nu_id_politician = nos_nu_corrected_only_politicians_no_duplicates[['dataset', 'id', 'true_label']]
```

```python
nos_nu_id_politician = nos_nu_id_politician.groupby(['id', 'dataset'], as_index=False).agg({'true_label': list})
```
