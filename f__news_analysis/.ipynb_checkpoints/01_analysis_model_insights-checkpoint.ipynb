```python
# c
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from collections import Counter

# Assume df and pol_img_only_ref_results are already defined as in your code
df = pol_img_only_ref_results.copy()
grouped = df.groupby('id')['per_politician'].apply(lambda x: list(set(x)))

co_occurrences = []
for politicians in grouped:
    if len(politicians) > 1:
        co_occurrences.extend(combinations(politicians, 2))
co_occurrence_counts = Counter(co_occurrences)

# Sort all_politicians alphabetically
all_politicians = [
    'Geert Wilders', 'Frans Timmermans', 'Dilan Yesilgöz', 'Pieter Omtzigt',
    'Rob Jetten', 'Caroline van der Plas', 'Henri Bontebal', 'Lilian Marijnissen',
    'Stephan van Baarle', 'Thierry Baudet', 'Esther Ouwehand', 'Mirjam Bikker',
    'Kees van der Staaij', 'Laurens Dassen', 'Joost Eerdmans', 'Edson Olf',
    'Wybren van Haga'
]

# Initialize confusion matrix
confusion_matrix = pd.DataFrame(0, index=all_politicians, columns=all_politicians)

# Fill the confusion matrix with co-occurrence counts
for (politician1, politician2), count in co_occurrence_counts.items():
    confusion_matrix.loc[politician1, politician2] = count
    confusion_matrix.loc[politician2, politician1] = count  # Symmetric matrix

# Convert DataFrame to NumPy array for manipulation
matrix_array = confusion_matrix.to_numpy()

# Mask the upper triangular part for lower triangle visualization
lower_triangular_array = np.tril(matrix_array)

# Create a mask for the heatmap: True where the upper triangular part is (including diagonal)
mask = np.triu(np.ones_like(lower_triangular_array, dtype=bool))

# Convert back to DataFrame for pretty printing
lower_triangular_df = pd.DataFrame(lower_triangular_array, index=confusion_matrix.index, columns=confusion_matrix.columns)

# Plot the lower triangular part using Seaborn heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    lower_triangular_df,
    annot=True,
    fmt='d',
    cmap='Oranges',
    mask=mask,
    cbar=True,
    square=True,
    vmin=0,
    vmax=lower_triangular_array.max()
)
plt.title('Co-occurrence Heatmap of Politicians in News Images (NOS, NU)\nOrdered by #seats after elections', fontdict=title_font)
plt.xlabel('Politician', fontdict=axis_font)
plt.ylabel('Politician', fontdict=axis_font)
plt.savefig('../z__visual-outputs-matplotlib/confusion_matrices_SVM_t1t2.pdf', format='pdf', bbox_inches='tight')

plt.show()
```

```python
# c Co-occurring visibility in main image
df = pol_img_only_ref_results.copy()
grouped = df.groupby('id')['per_politician'].apply(lambda x: list(set(x)))

co_occurrences = []
for politicians in grouped:
    if len(politicians) > 1:
        co_occurrences.extend(combinations(politicians, 2))
co_occurrence_counts = Counter(co_occurrences)

all_politicians = [
    'Geert Wilders',
    'Frans Timmermans',
    'Dilan Yesilgöz',
    'Pieter Omtzigt',
    'Rob Jetten',
    'Caroline van der Plas',
    'Henri Bontebal',
    'Lilian Marijnissen',
    'Stephan van Baarle',
    'Thierry Baudet',
    'Esther Ouwehand',
    'Mirjam Bikker',
    'Kees van der Staaij',
    'Laurens Dassen',
    'Joost Eerdmans',
    'Edson Olf',
    'Wybren van Haga'
]

confusion_matrix = pd.DataFrame(0, index=all_politicians, columns=all_politicians)
# Convert DataFrame to NumPy array for manipulation
matrix_array = confusion_matrix.to_numpy()

for (politician1, politician2), count in co_occurrence_counts.items():
    confusion_matrix.loc[politician1, politician2] = count
    confusion_matrix.loc[politician2, politician1] = count  # Symmetric matrix


# Mask the lower triangular part
upper_triangular_array = np.triu(matrix_array)

# Create a mask for the heatmap: True where upper triangular part is zero
mask = np.tril(np.ones_like(upper_triangular_array, dtype=bool))

# Convert back to DataFrame for pretty printing
upper_triangular_df = pd.DataFrame(upper_triangular_array, index=confusion_matrix.index, columns=confusion_matrix.columns)

# Plot the upper triangular part using Seaborn heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(upper_triangular_df, annot=True, fmt='d', cmap='Blues', mask=mask, cbar=True, square=True, vmin=0, vmax=upper_triangular_array.max())
plt.title('Upper Triangular Part of Co-Occurrence Matrix')
plt.xlabel('Politician')
plt.ylabel('Politician')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert DataFrame to NumPy array for manipulation
matrix_array = confusion_matrix.to_numpy()

# Mask the lower triangular part and diagonal
upper_triangular_array = np.triu(matrix_array)

# Create a mask that is True for lower triangular elements and the diagonal
mask = np.tril(np.ones_like(upper_triangular_array, dtype=bool))

# Convert back to DataFrame for pretty printing
upper_triangular_df = pd.DataFrame(upper_triangular_array, index=confusion_matrix.index, columns=confusion_matrix.columns)

# Plot the upper triangular part using Seaborn heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    upper_triangular_df,
    annot=True,
    fmt='d',
    cmap='Blues',
    mask=mask,
    cbar=True,
    square=True,
    vmin=0,
    vmax=upper_triangular_array.max(),
    linewidths=0.5,  # Optional: to add lines around cells for better visualization
    linecolor='white'  # Optional: line color for cells
)
plt.title('Upper Triangular Part of Co-Occurrence Matrix')
plt.xlabel('Politician')
plt.ylabel('Politician')
plt.show()
```

```python
def model_name(x):
    if x == 'svc_rbf':
        return 'SVM (rbf)'
    elif x == 'knn_10':
        return 'kNN (k=10)'
    elif x == 'knn_20':
        return 'kNN (k=20)'
    elif x == 'knn_30':
        return 'kNN (k=30)'
```

```python
df = pd.read_excel('datasets/input/model_selection.xlsx')
df['Model_name'] = df['Model'].apply(lambda x: model_name(x))
df_multiclass = df[df['Classification'] == 'Multiclass']
```

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-score']

colormap_mapping = [
    {'model': 'svc_rbf', 'name': 'SVM (rbf)', 'colormap': 'spring'},
    {'model': 'knn_10', 'name': 'kNN (k=10)', 'colormap': 'winter'},
    {'model': 'knn_20', 'name': 'kNN (k=20)', 'colormap': 'winter'},
    {'model': 'knn_30', 'name':'kNN (k=30)', 'colormap': 'winter'}
]

plt.figure(figsize=(14, 7))

x_labels = df_multiclass.apply(lambda row: f"{row['Model_name']}\nDimensionality={row['Num_dim']}\nThreshold={row['Threshold']}", axis=1)
x = np.arange(len(df_multiclass))
width = 0.2

for i, metric in enumerate(metrics):
    models = df_multiclass['Model'].values
    
    for j, model in enumerate(models):
        colormap_name = next((item['colormap'] for item in colormap_mapping if item['model'] == model), 'Greens')
        colormap = plt.colormaps.get_cmap(colormap_name)

        color = colormap(i / len(metrics))
        bars = plt.bar(x[j] + i * width, df_multiclass.iloc[j][metric], width, color=color, label=metric if j == 0 else "")

        yval = df_multiclass.iloc[j][metric]
        plt.text(
            x[j] + i * width,
            yval - 0.02, 
            round(yval, 2),
            ha='center',
            va='top',
            color='white',
            weight='bold',
            rotation=90
        )

plt.xlabel('Model, dimensionality, best performing threshold value', fontdict=axis_font)
plt.ylabel('Metric score', fontdict=axis_font)
plt.title('Comparison of models and parameter combinations,\nscores displayed for the threshold value with the best performance\n', fontdict=title_font)
plt.ylim(0, 1.09)
plt.xticks(x + width * 2, x_labels, rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
