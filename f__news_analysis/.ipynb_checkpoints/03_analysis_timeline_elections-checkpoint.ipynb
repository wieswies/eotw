### a. Patterns 
#### i. Total visibility in main image (per politician)

```python
total_per_politician_counts = pol_img_only_ref_results['per_politician'].value_counts().rename_axis('per_politician').reset_index(name='Combined')
nos_total_per_politician_counts = nos_pol_img_only_ref_results['per_politician'].value_counts().rename_axis('per_politician').reset_index(name='NOS')
nu_total_per_politician_counts = nu_pol_img_only_ref_results['per_politician'].value_counts().rename_axis('per_politician').reset_index(name='NU')

per_politician_count = nos_total_per_politician_counts.merge(nu_total_per_politician_counts, how='left')
per_politician_count = per_politician_count.merge(total_per_politician_counts, how='left')
per_politician_count['NU'] = per_politician_count['NU'].fillna(0).astype(int)
```

```python
df = per_politician_count.copy()
df.set_index('per_politician', inplace=True)
df = df.reindex(all_politicians)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, fmt='d', cmap='Oranges', cbar=True, linewidth=1, square=False, vmin=0, vmax=70)

# Customize titles and labels
plt.title("Visibility Heatmap: Party Leaders' Total Appearance\n", fontdict=title_font)
plt.xlabel('')
plt.ylabel('Politician', fontdict=axis_font)

plt.savefig('../z__visual-outputs-matplotlib/heatmap_total_visibility.pdf', format='pdf', bbox_inches='tight')
# Show the plot
plt.show()
```

#### ii. Single visibility in main image (per politician)

```python
total_single_politician = pol_img_only[pol_img_only['politician_in_img'].apply(lambda x:len(x) <= 1)]
total_single_politician_counts = total_single_politician['politician_in_img'].value_counts().rename_axis('per_single').reset_index(name='Combined')
total_single_politician_counts['per_single'] = total_single_politician_counts['per_single'].apply(lambda x: x[0])

nos_single_politician = nos_pol_img_only[nos_pol_img_only['politician_in_img'].apply(lambda x: len(x) <= 1)]
nos_single_politician_counts = nos_single_politician['politician_in_img'].value_counts().rename_axis('per_single').reset_index(name='NOS')
nos_single_politician_counts['per_single'] = nos_single_politician_counts['per_single'].apply(lambda x: x[0])

nu_single_politician = nu_pol_img_only[nu_pol_img_only['politician_in_img'].apply(lambda x: len(x) <= 1)]
nu_single_politician_counts = nu_single_politician['politician_in_img'].value_counts().rename_axis('per_single').reset_index(name='NU')
nu_single_politician_counts['per_single'] = nu_single_politician_counts['per_single'].apply(lambda x: x[0])

single_politician_count = nos_single_politician_counts.merge(nu_single_politician_counts, how='left')
single_politician_count = single_politician_count.merge(total_single_politician_counts, how='left')
single_politician_count['NU'] = single_politician_count['NU'].fillna(0).astype(int)
```

```python
if 'per_single' in single_politician_count.columns:
    single_politician_count.set_index('per_single', inplace=True)
df = single_politician_count.copy().reindex(all_politicians)

df['NOS'] = df['NOS'].fillna(0).astype(int)
df['NU'] = df['NU'].fillna(0).astype(int)
df['Combined'] = df['Combined'].fillna(0).astype(int)

plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, cmap='Oranges', cbar=True, linewidth=1, square=False, vmin=0, vmax=45)

plt.title("Visibility Heatmap: Party Leaders' Solo Appearence\n", fontdict=title_font)
plt.xlabel('')
plt.ylabel('Politician', fontdict=axis_font)

plt.savefig('../z__visual-outputs-matplotlib/heatmap_solo_visibility.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
# First
df1 = per_politician_count.copy()
df1.set_index('per_politician', inplace=True)
df1 = df1.reindex(all_politicians)

# Second 
if 'per_single' in single_politician_count.columns:
    single_politician_count.set_index('per_single', inplace=True)
df2 = single_politician_count.copy().reindex(all_politicians)

df2['NOS'] = df2['NOS'].fillna(0).astype(int)
df2['NU'] = df2['NU'].fillna(0).astype(int)
df2['Combined'] = df2['Combined'].fillna(0).astype(int)

# Subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

sns.heatmap(df1, annot=True, fmt='d', cmap='Oranges', cbar=True, linewidth=1, square=False, vmin=0, vmax=70, ax=ax1)
ax1.set_title("a. Party Leaders' Total Appearance\n", fontdict=title_font)
ax1.set_xlabel('')
ax1.set_ylabel('Politician', fontdict=axis_font)

sns.heatmap(df2, annot=True, cmap='Oranges', cbar=True, linewidth=1, square=False, vmin=0, vmax=70, ax=ax2)
ax2.set_title("b. Party Leaders' Solo Appearance\n", fontdict=title_font)
ax2.set_xlabel('')
ax2.set_ylabel('Politician', fontdict=axis_font)

plt.tight_layout()
plt.savefig('../z__visual-outputs-matplotlib/heatmap_combined_visibility.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
df3 = round((df2 / df1) * 100, 1)
df3 = df3.fillna(0)
df3 = df3.rename(columns={"per_single": "proportion_single"})

plt.figure(figsize=(10, 8))
sns.heatmap(df3, annot=True, fmt = '.1f', cmap='Oranges', cbar=True, linewidth=1, square=False, vmin=0, vmax=100)

plt.title("Visibility Heatmap: Proportion of Party Leaders Solo Appearance (Percentages)\n", fontdict=title_font)
plt.xlabel('')
plt.ylabel('Politician', fontdict=axis_font)

plt.savefig('../z__visual-outputs-matplotlib/heatmap_proportion_of_solo_visibility.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

#### iii. Images containing multiple party leaders in the image

```python
total_multi_politician = pol_img_only[pol_img_only['politician_in_img'].apply(lambda x: len(x) > 1)]
total_multi_politician_counts = total_multi_politician['politician_in_img'].value_counts().rename_axis('per_combo').reset_index(name='Combined')

nos_multi_politician = nos_pol_img_only[nos_pol_img_only['politician_in_img'].apply(lambda x: len(x) > 1)]
nos_multi_politician_counts = nos_multi_politician['politician_in_img'].value_counts().rename_axis('per_combo').reset_index(name='NOS')

nu_multi_politician = nu_pol_img_only[nu_pol_img_only['politician_in_img'].apply(lambda x: len(x) > 1)]
nu_multi_politician_counts = nu_multi_politician['politician_in_img'].value_counts().rename_axis('per_combo').reset_index(name='NU')

multi_counts = [total_multi_politician_counts, nos_multi_politician_counts, nu_multi_politician_counts]
for count in multi_counts:
    count['per_combo'] = count['per_combo'].astype(str)
    count['per_combo'] = count['per_combo'].str.replace('[', '').str.replace(']', '').str.replace("'", '')
```

```python
multi_politician_count = nos_multi_politician_counts.merge(nu_multi_politician_counts, how='left')
multi_politician_count = multi_politician_count.merge(total_multi_politician_counts, how='left')
multi_politician_count['NU'] = multi_politician_count['NU'].fillna(0).astype(int)
```

```python
df = multi_politician_count.copy()
df.set_index('per_combo', inplace=True)

df['NOS'] = df['NOS'].astype(int)
df['NU'] = df['NU'].astype(int)
df['Combined'] = df['Combined'].astype(int)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, fmt='d', cmap='Greens', cbar=True, linewidth=1, square=False, vmin=0, vmax=7)

plt.title("Party Leaders' Appearence in Combinations\n", fontdict=title_font)
plt.xlabel('')
plt.ylabel('Combination of politicians', fontdict=axis_font)

plt.savefig('../z__visual-outputs-matplotlib/heatmap_combi_visibility.pdf', format='pdf', bbox_inches='tight')
# Show the plot
plt.show()
```

#### iv. Co-occurrence matrix

```python
from itertools import combinations

# Initialize an empty co-occurrence matrix
co_occurrence_matrix = pd.DataFrame(0.0, index=all_politicians, columns=all_politicians)

for combo, combined_value in zip(df.index, df['Combined']):
    # Split the combination into individual politicians
    politicians_in_combo = combo.split(', ')
    
    # Generate all pairwise combinations
    for pol1, pol2 in combinations(politicians_in_combo, 2):
        if pol1 in all_politicians and pol2 in all_politicians:
            # Increment both (pol1, pol2) and (pol2, pol1) in the matrix by the 'Combined' value
            co_occurrence_matrix.loc[pol1, pol2] += combined_value
            co_occurrence_matrix.loc[pol2, pol1] += combined_value

co_occurrence_matrix.values[np.tril_indices_from(co_occurrence_matrix, k=0)] = np.nan

plt.figure(figsize=(10, 8))
ax = sns.heatmap(co_occurrence_matrix, annot=True, cmap='Purples', linewidths=0.5, square=True, 
                 xticklabels=True, yticklabels=True)

ax.xaxis.set_ticks_position('top')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='left')

ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.savefig('../z__visual-outputs-matplotlib/co_occurrence matrix.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
