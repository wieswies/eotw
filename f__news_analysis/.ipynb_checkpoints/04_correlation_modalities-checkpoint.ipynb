### b. Inspect textual patterns over time

```python
df = full_df.copy()
```

#### i. Pre-process

```python
df['date'] = pd.to_datetime(df['date'])
df['week'] = df['date'].dt.isocalendar().week

party_list = ['PVV', 'GL-PvdA', 'VVD', 'NSC', 'D66', 'BBB', 'CDA', 'SP', 'FVD', 'PvdD', 
              'CU', 'SGP', 'DENK', 'Volt', 'JA21', 'Bij1', 'BvNL']
count_df = pd.DataFrame()

# Counting
for party in party_list:
    df[f"{party}_paragraph_count"] = df[f"{party}_paragraphs"].apply(lambda x: len(x) if isinstance(x, list) else 0)
    
    # Group by week, sum
    weekly_counts = df.groupby('week')[f"{party}_paragraph_count"].sum().reset_index(name=f"{party}")
    
    # Merge results into count_df
    if count_df.empty:
        count_df = weekly_counts
    else:
        count_df = count_df.merge(weekly_counts, on='week', how='outer')
```

### d. Multi-modal correlations

```python
df = full_df.copy()
```

#### i. Pre-process

```python
df['date'] = pd.to_datetime(df['date'])
df['week'] = df['date'].dt.isocalendar().week
df['year'] = df['date'].dt.isocalendar().year
df['week_year'] = df['year'].astype(str) + '-' + df['week'].astype(str).str.zfill(2)
```

##### Visual data

```python
pol_img_ref_results_expanded = df.explode('politician_in_img')

pol_img_ref_results_expanded = pol_img_ref_results_expanded.dropna(subset=['politician_in_img'])

# Group by 'week_year' and 'politician_in_image' to get the count per week per politician
weekly_counts = pol_img_ref_results_expanded.groupby(['week_year', 'politician_in_img']).size().reset_index(name='count')

# Pivot the data to get the counts per politician per week in a matrix format suitable for plotting
visual_count_df = weekly_counts.pivot(index='week_year', columns='politician_in_img', values='count').fillna(0)
visual_count_df = visual_count_df.astype(int)
visual_count_df = visual_count_df.reindex(columns=all_politicians)
visual_count_df = visual_count_df.rename(columns=politician_party)
```

##### Textual data

```python
party_list = ['PVV', 'GL-PvdA', 'VVD', 'NSC', 'D66', 'BBB', 'CDA', 'SP', 'FVD', 'PvdD', 
              'CU', 'SGP', 'DENK', 'Volt', 'JA21', 'Bij1', 'BvNL']
count_df = pd.DataFrame()

for party in party_list:
    title_count = df[f"{party}_title"].apply(lambda x: len(x) if isinstance(x, list) else 0)
    paragraph_count = df[f"{party}_paragraphs"].apply(lambda x: len(x) if isinstance(x, list) else 0)

    df[f"{party}_title_text_count"] = ((2 * title_count) + paragraph_count)

# Melt the dataframe to have one row per party for easier grouping
melted_df = df.melt(id_vars=['week_year'], 
                         value_vars=[f"{party}_title_text_count" for party in party_list],
                         var_name='party', 
                         value_name='count')

melted_df['party'] = melted_df['party'].str.replace('_title_text_count', '')
weekly_counts = melted_df.groupby(['week_year', 'party']).sum().reset_index()

count_df = weekly_counts.pivot(index='week_year', columns='party', values='count').fillna(0)
tt_count_df = count_df.reindex(columns=party_list)
```

#### ii. Draw line-charts for different modalities

```python
textual_df = tt_count_df.copy()
visual_df = visual_count_df.copy()
```

```python
# Dual axis max values
max_textual = 750
max_visual = 35

num_parties = len(all_parties)

# Week numbers
week_numbers = textual_df.index.str.split('-').str[1]

# Subplots
fig, axes = plt.subplots(nrows=(num_parties // 3) + 1, ncols=3, figsize=(18, 24))
axes = axes.flatten() # Easy iteration

# Iterate through each party and create a dual-axis plot for each
for idx, party in enumerate(all_parties):
    ax = axes[idx]
    ax2 = ax.twinx()
    
    # Ax: Title and paragraph counts per party
    ax.plot(week_numbers, textual_df[party], color='cornflowerblue', label='Textual appearances')
    ax.set_ylim(0, max_textual)
    
    # Ax2: visual appearances of party leaders
    ax2.plot(week_numbers, visual_df[party], color='darkorange', label='Visual appearances')
    ax2.set_ylim(0, max_visual)
    
    # Set titles, labels, and legends
    ax.set_title(f'{party}', fontdict=title_font)
    ax.set_xlabel('Week')
    ax.set_ylabel('Textual appearances (title + text)', color='cornflowerblue', font=axis_font)
    ax2.set_ylabel('Visual appearances (party leader)', color='darkorange', font=axis_font)
    
    # Handle legends for each axis
    ax.legend(loc='upper left', bbox_to_anchor=(0, 1))
    ax2.legend(loc='upper left', bbox_to_anchor=(0, 0.9)) 

# Hide possible empty subplots (if any)
for ax in axes[num_parties:]:
    ax.set_visible(False)

# Adjust layout so the subplots don't overlap
plt.tight_layout()

plt.savefig('../z__visual-outputs-matplotlib/multi_modal_per_party.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])

df.set_index('date', inplace=True)

numeric_data = df[all_parties]

# Ensure only numeric columns 
numeric_data = numeric_data.select_dtypes(include='number')

# Normalize to correct for parties that left parliament
normalized_data = numeric_data.div(numeric_data.sum(axis=1), axis=0)

# Plotting a normalized stacked area chart for the granular data
plt.figure(figsize=(15, 8))

# Use the indexed numeric data for plotting
normalized_data.plot(kind='area', stacked=True, figsize=(15, 8), cmap='tab20')

plt.title('Normalized Peilingwijzer percentpoint estimates, averaged by week\nup to and including November 21, 2023\n', fontdict=title_font)
plt.xlabel('Week', fontdict=axis_font)
plt.ylabel('Peilingwijzier estimates', fontdict=axis_font)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Party leader', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

### ii. By week

```python
weekly_avg_df = df.groupby('week').mean().reset_index()
```

```python
# Ensure only numeric columns
numeric_data = weekly_avg_df.select_dtypes(include='number')

# Index by week for easier plotting
numeric_data.set_index('week', inplace=True)
numeric_data = numeric_data[all_parties]

# Normalize to correct for parties that left parliament
normalized_data = numeric_data.div(numeric_data.sum(axis=1), axis=0)

# Plotting a normalized stacked area chart for the averaged weekly values
plt.figure(figsize=(15, 8))

# Use the indexed numeric data for plotting
normalized_data.plot(kind='area', stacked=True, figsize=(15, 8), cmap='tab20')

plt.title('Normalized Peilingwijzer percentpoint estimates, averaged by week\nup to and including November 21, 2023\n', fontdict=title_font)
plt.xlabel('Week', fontdict=axis_font)
plt.ylabel('Peilingwijzier estimates', fontdict=axis_font)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Party leader', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
