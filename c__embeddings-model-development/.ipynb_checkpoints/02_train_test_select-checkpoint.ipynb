## 1. Data infrastructure

```python
embeddings = pd.read_parquet('../datasets/embeddings/embedded_politicians_pca_tsne.parquet')
```

```python
print(f'The dataset containing unique embeddings\nof both politicians and unknowns\nis of shape {embeddings.shape}\nand contains the columns:\n{embeddings.columns.tolist()}')
```

```python
label_encoder = LabelEncoder()
labels = embeddings['dataset']
y = label_encoder.fit_transform(labels)
embeddings['dataset_encoded'] = y
```

## 2. Model development
### a. Data preparation
#### i. Split the data into training (only politicians), development (politicians + unknowns), and testing (politicians + unknowns)

```python
pol_data = embeddings[embeddings['dataset'] != 'Unknown']
unknown_data = embeddings[embeddings['dataset'] == 'Unknown']

train_data, dev_data = train_test_split(pol_data, test_size=0.2, random_state=2)
dev_data = pd.concat([dev_data, unknown_data])
dev_data, test_data = train_test_split(dev_data, test_size=0.5, random_state=2)

dev_for_final_training = dev_data.copy()
additional_train_data = dev_for_final_training[dev_for_final_training['dataset'] != 'Unknown']
final_train_data = pd.concat([train_data, additional_train_data])

# Check the sizes of the splits
print(f"Training data size: {train_data.shape}")
print(f"Development data size: {dev_data.shape}")
print(f"Final training data size: {final_train_data.shape}")
print(f"Test data size: {test_data.shape}")
```

#### ii. Set data dimensions

```python
# Set X_train, y__train, X_dev, y_dev, X_test, y_test for the different data dimensionalities
X_train = np.vstack(train_data['embedding'].values)
X_train_PCA = np.vstack(train_data['PCA_64D'].values)
X_train_tSNE = np.vstack(train_data['tSNE_2D'].values)
y_train = train_data['dataset_encoded']

X_dev = np.vstack(dev_data['embedding'].values)
X_dev_PCA = np.vstack(dev_data['PCA_64D'].values)
X_dev_tSNE = np.vstack(dev_data['tSNE_2D'].values)
y_dev = dev_data['dataset_encoded']

X_train_final = np.vstack(final_train_data['embedding'].values)
X_train_PCA_final = np.vstack(final_train_data['PCA_64D'].values)
X_train_tSNE_final = np.vstack(final_train_data['tSNE_2D'].values)
y_train_final = final_train_data['dataset_encoded']

X_test = np.vstack(test_data['embedding'].values)
X_test_PCA = np.vstack(test_data['PCA_64D'].values)
X_test_tSNE = np.vstack(test_data['tSNE_2D'].values)
y_test = test_data['dataset_encoded']
```

```python
print(f'Classes in the training set: {y_train.nunique()} ({sorted(y_train.unique())})\n')
print(f'Classes in the dev set: {y_dev.nunique()} ({sorted(y_dev.unique())})\n')
print(f'Classes in the final training set: {y_train_final.nunique()} ({sorted(y_train_final.unique())})\n')
print(f'Classes in the test set: {y_test.nunique()} ({sorted(y_test.unique())})\n')
```

### b. Create pipeline
#### i. Create libraries to facilitate looping

```python
X_data = [
    {
        'num_dim': 512,
        'X_data': {
            'X_train': X_train,
            'X_dev': X_dev,
            'X_train_final': X_train_final,
            'X_test': X_test
        }
    },
    {
        'num_dim': 64,
        'X_data': {
            'X_train': X_train_PCA,
            'X_dev': X_dev_PCA,
            'X_train_final': X_train_PCA_final,
            'X_test': X_test_PCA
        }
    },
    {
        'num_dim': 2,
        'X_data': {
            'X_train': X_train_tSNE,
            'X_dev': X_dev_tSNE,
            'X_train_final': X_train_tSNE_final,
            'X_test': X_test_tSNE
        }
    }
]

y_data = [
    {
        'y_train': y_train,
        'y_dev': y_dev,
        'y_train_final': y_train_final,
        'y_test': y_test
    }
]

thresholds = [
    {
        't_name': 't1',
        't_value': 0.4
    },
    {
        't_name': 't2',
        't_value': 0.5
    },
    {
        't_name': 't3',
        't_value': 0.6
    },
    {
        't_name': 't4',
        't_value': 0.7
    },
    {
        't_name': 't5',
        't_value': 0.8
    }
]
```

#### ii. Define data pipelines for robust testing

```python
pipe_svc = Pipeline([('scaler', StandardScaler()), ('svc_rbf', SVC(probability=True))])
pipe_knn_10 = Pipeline([('scaler', StandardScaler()), ('knn_10', KNeighborsClassifier(n_neighbors=10))])
pipe_knn_20 = Pipeline([('scaler', StandardScaler()), ('knn_20', KNeighborsClassifier(n_neighbors=20))])
pipe_knn_30 = Pipeline([('scaler', StandardScaler()), ('knn_30', KNeighborsClassifier(n_neighbors=30))])
```

#### iii. Ready, set, train!

```python
results_svc = train_dev_dimensions(pipe_svc, X_data, y_data, thresholds)
results_knn_10 = train_dev_dimensions(pipe_knn_10, X_data, y_data, thresholds)
results_knn_20 = train_dev_dimensions(pipe_knn_20, X_data, y_data, thresholds)
results_knn_30 = train_dev_dimensions(pipe_knn_30, X_data, y_data, thresholds)
```

```python
model_performance_svc = pd.DataFrame(results_svc)
model_performance_knn_10 = pd.DataFrame(results_knn_10)
model_performance_knn_20 = pd.DataFrame(results_knn_20)
model_performance_knn_30 = pd.DataFrame(results_knn_30)
```

```python
model_performance = pd.concat([model_performance_svc, model_performance_knn_10, model_performance_knn_20, model_performance_knn_30], 
                              ignore_index=True)
```

```python
model_performance.to_excel('dev_metrics.xlsx')
```

#### iv. Testing

```python
test_results_svc = test_dimensions(pipe_svc, X_data, y_data, thresholds)
test_results_knn_10 = test_dimensions(pipe_knn_10, X_data, y_data, thresholds)
test_results_knn_20 = test_dimensions(pipe_knn_20, X_data, y_data, thresholds)
test_results_knn_30 = test_dimensions(pipe_knn_30, X_data, y_data, thresholds)
```

```python
test_performance_svc = pd.DataFrame(test_results_svc)
test_performance_knn_10 = pd.DataFrame(test_results_knn_10)
test_performance_knn_20 = pd.DataFrame(test_results_knn_20)
test_performance_knn_30 = pd.DataFrame(test_results_knn_30)
```

```python
test_performance = pd.concat([test_performance_svc, test_performance_knn_10, test_performance_knn_20, test_performance_knn_30], 
                              ignore_index=True)
```

```python
test_performance.to_excel('test_metrics.xlsx')
```

#### v. Merge development model performances and test model performances to compare accuracies

```python
dev_model_performances = [model_performance_svc, model_performance_knn_10, model_performance_knn_20, model_performance_knn_30]
test_model_performances = [test_performance_svc, test_performance_knn_10, test_performance_knn_20, test_performance_knn_30]
model_names = ['SVC (rbf)', 'kNN (n=10)', 'kNN(n=20)', 'kNN(n=30)']
```

```python
for model_performance, model_name in (zip(test_model_performances, model_names)):
    print(f"Metrics for {model_name}")
    df = model_performance.copy()
    df_binary = df[df['Classification'] == 'Binary']
    df_multiclass = df[df['Classification'] == 'Multiclass']

    plot_metrics(df_binary, f'{model_name} - Binary Classification Metrics')
    plot_metrics(df_multiclass, f'{model_name} - Multiclass Classification Metrics')
```

## 3. Model selection

```python
test_performance[(test_performance['Classification'] == 'Multiclass') & (test_performance['Model'] == 'svc_rbf') & (test_performance['Num_dim'] == 512)]
```

Based on the metrics report, and evaluation of the confusion matrix, the following model is chosen: 
###### SVM (kernel=rbf), num_dim = 512, threshold = 0.7
Multiclass performance:
- Accuracy: 0.94
- Precision: 0.91
- Recall: 0.99
- F1-score: 0.94

Although the metrics look slightly better for threshold values 0.6 and 0.7, the cost of mistaking politicians for unknowns is higher, hence the model is chosen with a decent overall performance and 0 missed politicians on the test set.
