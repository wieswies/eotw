#### Copy unique images in preparation of data augmentation
Use 'z' to 'walk over' all unique photos by filename and write them into an output folder such that the unique photos can be augmented for resolving class imbalance.

```python
input_path = Path('../Datasets/Training_in_isolation')
output_path = Path('../Datasets/Data_augmented')
unique_embeddings = [str(item) for item in embeddings['z']]
```

```python
for politician_subfolder in os.listdir(input_path):
    # Loop over all subfolders of politicians
    folder_to_split = os.path.join(input_path, politician_subfolder)
    destination_folder = os.path.join(output_path, politician_subfolder)
    '''
    run once:
    os.makedirs(destination_folder, exist_ok = True)
    '''
    
    print(f'Splitting unique images of politician {politician_subfolder}')
    
    for file in os.listdir(folder_to_split):
        # Loop over all files in the politician's subfolder and check if
        # they occur in your list with unique embeddings
        if file in unique_embeddings:
            # Copy unique files to data augmentation folder
            origin_path = os.path.join(folder_to_split, file)
            destination_path = os.path.join(destination_folder, file)
            shutil.copy(origin_path, destination_path)
    print(f'Copied unique images of politician {politician_subfolder} to {destination_folder}')
```

#### Data augmentation
Additional rotation for minority classes: Joost Eerdmans, Wybren van Haga, Laurens Dassen, Stephan an Baarle (2x), Edson Olf (2x)

```python
minority_classes = [
    'Joost Eerdmans',
    'Wybren van Haga',
    'Laurens Dassen',
    'Stephen van Baarle',
    'Edson Olf'
]
strict_minorities = [
    'Stephan van Baarle',
    'Edson Olf'
]
```

###### Overall flip for extra robustness

```python
for politician_subfolder in os.listdir(output_path):
    politician_to_flip = os.path.join(output_path, politician_subfolder)
    
    # Loop through image items for selected politician
    for i, file in enumerate(os.listdir(politician_to_flip)):
        
        file_path = os.path.join(politician_to_flip, file)
        politician_image = Image.open(file_path)
        
        # Transpose image
        flipped_politician_image = politician_image.transpose(Image.FLIP_LEFT_RIGHT)
        
        flippy_file_path = os.path.join(politician_to_flip, 'flipped_' + file)
        flipped_politician_image.save(flippy_file_path)
```

###### Small rotation for minority classes

```python
for politician_subfolder in os.listdir(output_path):
    if politician_subfolder in minority_classes:
        
        politician_to_rotate = os.path.join(output_path, politician_subfolder)

        # Loop through image items for selected politician
        for i, file in enumerate(os.listdir(politician_to_rotate)):

            file_path = os.path.join(politician_to_rotate, file)
            politician_image = Image.open(file_path)

            # Rotate image
            rotated_politician_image = politician_image.rotate(25)

            rotated_file_path = os.path.join(politician_to_rotate, 'rotated_' + file)
            rotated_politician_image.save(rotated_file_path)
```

###### Distortion for strict minorities

```python
for politician_subfolder in os.listdir(output_path):
    if politician_subfolder in strict_minorities:
        
        politician_to_distort = os.path.join(output_path, politician_subfolder)

        # Loop through image items for selected politician
        for i, file in enumerate(os.listdir(politician_to_distort)):

            file_path = os.path.join(politician_to_distort, file)
            politician_image = Image.open(file_path)

            # Rotate image
            distorted_politician_image = politician_image.filter(ImageFilter.UnsharpMask(radius=3, percent=200, threshold=5))

            distorted_file_path = os.path.join(politician_to_distort, 'distorted_' + file)
            distorted_politician_image.save(distorted_file_path)
```

### Finally: calculate embeddings for unique, flipped, rotated and distorted images

```python
class HiddenPrints:
    def __enter__(self):
        self._original_stdout = sys.stdout
        sys.stdout = open(os.devnull, 'w')

    def __exit__(self, exc_type, exc_val, exc_tb):
        sys.stdout.close()
        sys.stdout = self._original_stdout
```

```python
# Method to extract the face embedding of an image
def get_face_embedding(face_image):
    detections = embedder.extract(face_image, threshold=0.95)
    
    # Check if a face was found
    if len(detections) > 0:
        # Return the embedding of the found face
        return detections[0]['embedding']
    else: return None

    time.sleep(0.5)
```

```python
data_folder = Path('../Datasets/Data_augmented')
```

```python
for folder in os.listdir(data_folder):
    print(folder)
```

```python
A, b, c = [], [], []
empty_embeddings = []
```

```python
embedder = FN()
```

```python
# Loop over politician subfolders in the isolated face image data folder
for politician_subfolder in tqdm(os.listdir(data_folder), desc='Embedding, embedding, embedding aan de wand', unit='politician_subfolder'):
    if politician_subfolder.startswith('.'):
        continue
    politician_path = os.path.join(data_folder, politician_subfolder)

    # Loop over the image files in the directory
    for iso_img_file in os.listdir(politician_path):
        # Check if file is an image
        if not iso_img_file.endswith('.jpeg'):
                continue
        iso_img_path = os.path.join(politician_path, iso_img_file)

        # Read the image files and extract the face_embeddings with cv2
        if cv2.imread(iso_img_path) is not None:
            with HiddenPrints():
                # Call method to extract the face embedding
                face_embedding = get_face_embedding(cv2.imread(iso_img_path))

                # Leave out None-embedded images
                if face_embedding is not None:
                    A.append(face_embedding)
                    b.append(politician_subfolder)
                    c.append(iso_img_file)
                else:
                    empty_embeddings.append((politician_subfolder, iso_img_file))
    time.sleep(10)
```

```python
print(f'Total number of calculated embeddings: {len(A)} \nTotal number of empty embeddings: {len(empty_embeddings)}')
```

```python
npX = np.array(A)
npy = np.array(b)
npz = np.array(c)
```

```python
embedded_data = pd.DataFrame({'X': npX.tolist(), 'y': npy, 'Filename': npz})
embedded_data.to_excel('../Datasets/Embeddings_labels_augmented_final.xlsx')
```
